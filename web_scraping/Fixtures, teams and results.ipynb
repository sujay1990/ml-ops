```python
import os 
import pandas as pd
import numpy as np
```

```python
os.chdir("F:\\Road to Opta\\player_stats\\")
football_dataset = pd.read_csv("football_dataset.csv")
```

```python
football_dataset = football_dataset.iloc[:,1:]
```

```python
football_dataset.columns = map(str.lower, football_dataset.columns)
```

```python
football_dataset.rename(columns={'player name': 'player_name'},inplace=True)
```

```python
columns = football_dataset.columns.tolist()
```

```python
new_columns = columns[:1] + columns[-1:] + columns[1:-1]
```

```python
football_dataset = football_dataset[new_columns]
```

```python
football_dataset.sort_values(['season', 'league', 'team', 'opp','h/a'], inplace=True)
```

```python
football_dataset.to_csv("football_dataset_sorted.csv")
```

```python
football_dataset = pd.read_csv("football_dataset_sorted.csv")
```

```python
football_dataset = football_dataset.iloc[:,1:]
football_dataset['team_goals'] = football_dataset.groupby(['season','team','league','opp','h/a'])['g'].transform('sum')
football_dataset['own_goals'] = football_dataset.groupby(['season','team','league','opp','h/a'])['own'].transform('sum')
```

```python
football_dataset_clone = football_dataset
```

```python
idx = football_dataset_clone.index
football_dataset_clone.loc[idx,['team','opp']] = football_dataset_clone.loc[idx,['opp','team']].values
football_dataset_clone['h/a'].replace({"A": "H", "H": "A"}, inplace=True)
```

```python
football_dataset = pd.read_csv("football_dataset_sorted.csv")
football_dataset = football_dataset.iloc[:,1:]
football_dataset['team_goals'] = football_dataset.groupby(['season','team','league','opp','h/a'])['g'].transform('sum')
football_dataset['own_goals'] = football_dataset.groupby(['season','team','league','opp','h/a'])['own'].transform('sum')
```

```python
football_dataset_clone.sort_values(['season', 'league', 'team', 'opp','h/a'], inplace=True)
```

```python
football_dataset['opp_goals'] = football_dataset_clone['team_goals'].values
football_dataset['opp_own_goals'] = football_dataset_clone['own_goals'].values
```

```python
football_dataset['team_score'] = football_dataset['team_goals'] + football_dataset['opp_own_goals']
football_dataset['opp_score'] = football_dataset['opp_goals'] + football_dataset['own_goals']
```

```python
football_dataset.drop(['team_goals','opp_goals','opp_own_goals','own_goals'],axis=1)
```

```python
football_dataset.to_csv("football_dataset_scoreline.csv")
```
