```python
import os 
import pandas as pd
import numpy as np
os.chdir("F:\\Road to Opta\\player_stats\\")
football_dataset = pd.read_csv("football_dataset_sorted.csv")
football_dataset.sort_values(['season', 'team', 'league', 'opp','h/a'], inplace=True)
football_dataset.dropna(inplace=True)
football_dataset = football_dataset.iloc[:,1:]
football_dataset['team_goals'] = football_dataset.groupby(['season','team','league','opp','h/a'])['g'].transform('sum')
football_dataset['own_goals'] = football_dataset.groupby(['season','team','league','opp','h/a'])['own'].transform('sum')
```

```python
football_dataset['opp_goals'] = football_dataset_clone['team_goals'].values
football_dataset['opp_own_goals'] = football_dataset_clone['own_goals'].values
```

```python
football_dataset.tail(38)
```

```python
football_dataset['team_score'] = football_dataset['team_goals'] + football_dataset['opp_own_goals']
football_dataset['opp_score'] = football_dataset['opp_goals'] + football_dataset['own_goals']
```

```python
football_dataset.drop(['team_goals','opp_goals','opp_own_goals','own_goals'],axis=1)
```

```python
conditions = [
    (football_dataset['team_score'] > football_dataset['opp_score']),
    (football_dataset['team_score'] < football_dataset['opp_score']),
    (football_dataset['team_score'] == football_dataset['opp_score'])
    ]

# create a list of the values we want to assign for each condition
values = ['win', 'loss', 'draw']

# create a new column and use np.select to assign values to it using our lists as arguments
football_dataset['result'] = np.select(conditions, values)

# display updated DataFrame
football_dataset.head()
```

```python
football_dataset.to_csv("football_dataset_scoreline.csv")
```
