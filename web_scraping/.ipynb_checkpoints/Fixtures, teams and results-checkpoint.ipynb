```python
import os 
import pandas as pd
import numpy as np
```

```python
os.chdir("F:\\Road to Opta\\player_stats\\")
football_dataset = pd.read_csv("football_dataset.csv")
```

```python
football_dataset = football_dataset.iloc[:,1:]
```

```python
football_dataset.columns = map(str.lower, football_dataset.columns)
```

```python
football_dataset.rename(columns={'player name': 'player_name'},inplace=True)
```

```python
columns = football_dataset.columns.tolist()
```

```python
new_columns = columns[:1] + columns[-1:] + columns[1:-1]
```

```python
football_dataset = football_dataset[new_columns]
```

```python
football_dataset.sort_values(['season', 'league', 'team', 'opp','h/a'], inplace=True)
```

```python
football_dataset.to_csv("football_dataset_sorted.csv")
```

```python
test_agg = pd.read_csv("test_agg.csv")
```

```python
sum_df = test_agg.groupby(['season','league','team','opp','h/a']).agg({'g': 'sum','own':'sum'})
```

```python
test_agg['team_goals'] = test_agg.groupby(['season','team','league','opp','h/a'])['g'].transform('sum')
test_agg['own_goals'] = test_agg.groupby(['season','team','league','opp','h/a'])['own'].transform('sum')
```

```python
test_agg_2 = test_agg[['season','team','league','opp','h/a','team_goals','own_goals']]
```

```python
test_agg_2 = test_agg_2.drop_duplicates().reset_index(drop=True)
```

```python
for index, row in test_agg.iterrows():
    for i, r in test_agg_2.iterrows():
        if ((row[1] == r[0]) & (row[2]==r[3]) & \
                                      (row[3] == r[2]) & (row[4]==r[1]) & \
                                        (row[5] != r[4])):
#             print("index", index, "team_goals", row[-2], "opp goals", r[5], "own_goals", r[6])
            test_agg.loc[index,"opp_goals"] = r[5]
    
            #test_agg.loc[index,"own_goals"] = row[53]
            if r[6]>0:
                test_agg.loc[index,'team_goals'] = test_agg.loc[index,'team_goals'] + r[6]
test_agg["opp_goals"] = test_agg["opp_goals"] + test_agg["own_goals"]
```

```python
test_agg.to_csv("test_agg_opp_goals.csv")
```
